```python
import matplotlib.pyplot as plt
import numpy as np
from imagep import gui
```

```python
origin = gui('./test.png', color='white')[0]
```

```python
origin
```

```python
xcal = gui('./test.png', origin=origin, color='black')[:,0]
```

```python
plt.figure()
# Calculate differences so we can calculate the div
xdiff = [xcal[i+1]-xcal[i] for i in range(len(xcal)-1)]
xdiff_mean = np.mean(xdiff)
plt.plot(xdiff)
plt.plot([0, len(xdiff)-1], np.ones(2)*xdiff_mean, label='mean')
plt.legend()
plt.grid()
```

```python
xpixel = 0.25/xdiff_mean # cm
print(xpixel)
```

```python
ycal = gui('./test.png', origin=origin, color='white')[:,1]
```

```python
plt.figure()
# Calculate differences so we can calculate the div
ydiff = [ycal[i+1]-ycal[i] for i in range(len(ycal)-1)]
ydiff_mean = np.mean(ydiff)
plt.plot(ydiff)
plt.plot([0, len(ydiff)-1], np.ones(2)*ydiff_mean, label='mean')
plt.legend()
plt.grid()
```

```python
ypixel = 0.25/35.5 # cm
```

```python
gui('./test.png', origin=origin, calibration=(xpixel, ypixel), unit='cm', color='black')
```
